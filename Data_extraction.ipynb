```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from PIL import Image

# Configuración
dataset_path = "/content/drive/MyDrive/CocaVsPepsi/dataset/coca/Lata De coca cola regular"  # Carpeta con las imágenes descargadas
output_path = "/content/drive/MyDrive/CocaVsPepsi/processed_dataset/coca"  # Carpeta donde guardaremos las imágenes normalizadas
img_size = (512, 512)  # Tamaño deseado para las imágenes

# Crear carpetas de salida
train_path = os.path.join(output_path, "train-coca")
val_path = os.path.join(output_path, "val-coca")
test_path = os.path.join(output_path, "test-coca")
for path in [train_path, val_path, test_path]:
    os.makedirs(path, exist_ok=True)

# Leer imágenes y preprocesarlas
images = []
image_names = []
for file in os.listdir(dataset_path):
    img_path = os.path.join(dataset_path, file)
    try:
        img = cv2.imread(img_path)
        img = cv2.resize(img, img_size)  # Redimensionar
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convertir a RGB
        img = img / 255.0  # Normalizar (valores entre 0 y 1)
        images.append(img)
        image_names.append(file)
    except:
        print(f"Error al procesar {file}")

# Convertir a numpy array
images = np.array(images)

# Dividir en conjuntos (80% entrenamiento, 10% validación, 10% prueba)
train_imgs, test_imgs, train_names, test_names = train_test_split(images, image_names, test_size=0.2, random_state=42)
val_imgs, test_imgs, val_names, test_names = train_test_split(test_imgs, test_names, test_size=0.5, random_state=42)

# Guardar imágenes preprocesadas
def save_images(img_list, name_list, folder):
    for img, name in zip(img_list, name_list):
        img = (img * 255).astype(np.uint8)  # Convertir de nuevo a 8-bit
        img = Image.fromarray(img)
        img.save(os.path.join(folder, name))

save_images(train_imgs, train_names, train_path)
save_images(val_imgs, val_names, val_path)
save_images(test_imgs, test_names, test_path)

print("Imágenes procesadas y organizadas en conjuntos de entrenamiento, validación y prueba.")
```
